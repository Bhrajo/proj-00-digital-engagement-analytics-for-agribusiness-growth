### **DATA CLEANING AND PREPROCESSING PROJECT** <br> *Project Level: Beginner <br> Data Source: 365-DataScience Absenteeism-Data*
#### **OBJECTIVE:**<br>A reputable firm has collected a fairly cleaned data on their employees' absence from work during normal working hours, resulting <br>in temporary incapacity to execute regular working activities. As a Data Analyst, I am tasked to ensure that the raw dataset, as was <br> collected, has been turned into a meaningful quantitative information for modelling and prescriptive analytics in Python.

#### **Importing Libraries and Loading Dataset**

```python
import numpy as np
import pandas as pd
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
raw_data =  pd.read_csv("Absenteeism-data.csv")

df = raw_data.copy()
print(f"Shape of the DataFrame:\n{df.shape}\n")
df.head()
```

#### **Inspecting Dataset**

```python
print(df.info(), '\n')

df.describe()
```

#### **ID**

```python
## Dropping the nominal column [ID] to make room for direct overarching quatitative analysis
df = df.drop(['ID'], axis = 1)
df.head()
```

#### **Reason for Absence**

```python
## Column Values:
print(f"Reason for Absence:\n{df['Reason for Absence'].tolist()}\n")
## Min & Max Values
print(f"Max_Value:\n{df['Reason for Absence'].max()}\n\nMin_Value:\n{df['Reason for Absence'].min()}\n")
## Unique Values
print(f"Sorted Unique Values:\n{np.sort(pd.unique(df['Reason for Absence']))}\n\nNumber of Unique Values:\n{df['Reason for Absence'].nunique()}")
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], dtype= int)
reason_columns.head(10)
```

```python
## Creating a [check] column to check the data integrity of the dummy variables 
reason_columns['check'] = reason_columns.sum(axis = 1)
reason_columns.head(10)
```

```python
## Checking the uniqueness of the dummy variables
print(pd.unique(reason_columns.check))
reason_columns['check'].sum(axis = 0)
```

```python
## Dropping the [check] column after validating the data integrity of the dummy variables
reason_columns = reason_columns.drop(['check'], axis = 1)

reason_columns.head(10)
```

```python
## Dropping the first column of the dummy variable (all zeroes) to avoid potential multicollinearity
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
reason_columns.head(10)
```

#### **Grouping Reasons for Absence**

```python
print(df.columns.values, '\n')
print(reason_columns.columns.values)
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
df.head(10)
```

```python
## Generating groups from the 'reason_columns'
reason_type_1 = reason_columns.loc[:,1:14].max(axis = 1).astype(int)
reason_type_2 = reason_columns.loc[:,15:17].max(axis = 1).astype(int)
reason_type_3 = reason_columns.loc[:,18:21].max(axis = 1).astype(int)
reason_type_4 = reason_columns.loc[:,22:].max(axis = 1).astype(int)

reason_type_1
```

#### **Concatenate Column Values**

```python
## Adding the [Reason for Absence] column back to the dataframe
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df.head(10)
```

```python
## Checking new column names 
df.columns.values
```

```python
## Changing column names to make [0,1,2,3] more meaningful
columns_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
                  'Daily Work Load Average', 'Body Mass Index', 'Education',
                  'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 
                  'Reason_2', 'Reason_3', 'Reason_4']

df.columns  = columns_names
df.head()
```

#### **Reorder Columns in DataFrame**

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 
                          'Distance to Work', 'Age','Daily Work Load Average', 'Body Mass Index', 
                          'Education','Children', 'Pets', 'Absenteeism Time in Hours']

df = df[column_names_reordered]
df.head()
```

```python
## Checkpoint
df_reason_mode = df.copy()
df_reason_mode.head(10)
```

#### **Date**

```python
df_reason_mode['Date'].head(10)
```

```python
## Checking original datatype of the [Date] column
print(type(df_reason_mode['Date']))
print(df_reason_mode.Date.dtype)
print(type(df_reason_mode['Date'][0]))
print(type(df_reason_mode.Date[0]))

## Converting the [Date] column from 'str' to a timestamp,'pd.to_datetime()
df_reason_mode['Date'] = pd.to_datetime(df_reason_mode['Date'], format = '%d/%m/%Y')

## Checking data type after conversion
print(type(df_reason_mode.Date[0])) 
df_reason_mode.Date
```

#### **Extracting Month Value**

```python
df_reason_mode.Date[0]
```

```python
## Extracting Year Value
print(df_reason_mode.Date[0].year)

## Extracting Month Value
print(df_reason_mode.Date[0].month)

## Extracting Day Value
print(df_reason_mode.Date[0].weekday())
```

```python
day_list = []
for i in range(df_reason_mode.shape[0]):
    day_list.append(df_reason_mode.Date[i].weekday())
df_reason_mode['Day of the Week'] = day_list
```

```python
## Creating Day Value column
def date_to_weekday(date_value):
    return date_value.weekday()
df_reason_mode['Date'].apply(date_to_weekday)

## Creating a Month Value column 
months_list = []
for i in range(df_reason_mode.shape[0]):
    months_list.append(df_reason_mode.Date[i].month)
df_reason_mode['Month Value'] = months_list

## Creating a Year Value column
year_list =  []
for i in range(df_reason_mode.shape[0]):
    year_list.append(df_reason_mode.Date[i].year)
df_reason_mode['Year Value'] = year_list

df_reason_mode.head(20)
```

```python
df_reason_mode.columns.values
```

```python
new_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Day of the Week',
               'Month Value','Transportation Expense', 'Distance to Work', 'Age',
               'Daily Work Load Average', 'Body Mass Index', 'Education',
               'Children', 'Pets', 'Absenteeism Time in Hours']

df_reason_mode = df_reason_mode[new_columns]
df_reason_mode.head()
```

#### **The The Data Types of the Next Five Relevant Columns**

```python
print(type(df_reason_mode['Transportation Expense'][0]))
print(type(df_reason_mode['Distance to Work'][0]))
print(type(df_reason_mode['Age'][0]))
print(type(df_reason_mode['Daily Work Load Average'][0]))
print(type(df_reason_mode['Body Mass Index'][0]))
```

#### **'Education', 'Children', 'Pets'**

```python
for i in df_reason_mode.loc[:,'Education':'Pets']:
    ## Unique values
    print(f"Unique Values in {i}:\n{df_reason_mode[i].unique()}")
    ## Number of unique values
    print(f"Number of Unique Values in {i}:\n{df_reason_mode[i].nunique()}")
    ## Unique Value Count
    print(f"Unique Value Counts for {i}:\n{df_reason_mode[i].value_counts()}\n")
```

```python
## Generating 2 categories for Unique Value Counts for [Education]: {1, 2, 3, 4. 1:0 and 2,3,4:1}
df_reason_mode['Education'] = df_reason_mode.Education.map({1:0, 2:1, 3:1, 4:1})
print(f"---Education---:\nUnique Values: {df_reason_mode.Education.unique()}\n\nUnique Value Counts:\n{df_reason_mode.Education.value_counts()} ")
```
