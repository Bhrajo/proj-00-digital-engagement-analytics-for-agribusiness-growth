```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### **Bar Charts**

```python
df_used_cars =pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df_used_cars["Brand"],
        height = df_used_cars["Cars Listings"], color = "midnightblue")
plt.xticks(rotation = -45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Cars Listing by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)
plt.show
```

#### **Pie Charts**

```python
fuel_engine_type_df = pd.read_csv("pie_chart_data.csv")
fuel_engine_type_df
```

```python
plt.figure(figsize = (10,8))
plt.pie(fuel_engine_type_df["Number of Cars"], 
        labels = fuel_engine_type_df["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "x-large",
                     "fontweight": "bold",
                     "rotation": 45,
                     'color': 'w'})
plt.legend(loc = "lower left")
plt.title('Cars by Engine Fuel Type', fontsize = 18, fontweight = 'bold')
plt.show()
```

#### **Stack Area Chart**

```python
df_fuel_eng_types = pd.read_csv("stacked_area_chart_data.csv")
df_fuel_eng_types.head()
```

```python
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Diesel', 'Petrol', 'Gas']
sns.set_style ('white')
plt.figure(figsize = (15,6))
plt.stackplot(df_fuel_eng_types["Year"],
              df_fuel_eng_types["Diesel"],
              df_fuel_eng_types["Petrol"],
              df_fuel_eng_types["Gas"],
              colors = colors,
              edgecolor = 'None')
plt.xticks(df_fuel_eng_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13, fontweight = 'bold')
plt.xlabel("Year", fontsize = 13, fontweight = 'bold')
plt.title("Popularity of Engine Fuel Types(1982 - 2016)", fontsize = 16, fontweight = 'bold')
sns.despine()
plt.show()
```

#### **Line Chart**

```python
df_spx_ftse = pd.read_csv("line_chart_data.csv")
df_spx_ftse.head(11)
```

```python
df_spx_ftse['New_Date'] = pd.to_datetime(df_spx_ftse.Date)
df_spx_ftse
```

```python
sns.set()
```

```python
legend_labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20, 8))
plt.plot(df_spx_ftse['New_Date'], df_spx_ftse['GSPC500'])
plt.plot(df_spx_ftse['New_Date'], df_spx_ftse['FTSE100'])
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize  = 16, fontweight = 'bold')
plt.ylabel("Returns", fontsize = 13, fontweight = 'bold')
plt.xlabel("Date", fontsize = 13, fontweight = 'bold')
plt.legend(labels = legend_labels,  fontsize = 'large')

plt.show()
```

```python
df_spx_ftse_H2_08 = df_spx_ftse[(df_spx_ftse.New_Date >= '2008-07-01') & (df_spx_ftse.New_Date <= '2008-12-31')]
df_spx_ftse_H2_08
```

```python
##legend_labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20, 8))
plt.plot(df_spx_ftse_H2_08['New_Date'], df_spx_ftse_H2_08['GSPC500'], color = 'midnightblue')
plt.plot(df_spx_ftse_H2_08['New_Date'], df_spx_ftse_H2_08['FTSE100'], color = 'crimson')
plt.title("S&P vs FTSE Returns (H2 2008)", fontsize  = 16, fontweight = 'bold')
plt.ylabel("Returns", fontsize = 13, fontweight = 'bold')
plt.xlabel("Date", fontsize = 13, fontweight = 'bold')
plt.legend(labels = legend_labels,  fontsize = 'large')

plt.show()
```

#### **Histograms**

```python
real_estate_df = pd.read_csv("histogram_data.csv")
real_estate_df
```

```python
sns.set_style('white')
plt.figure(figsize = (8,6))
plt.hist(real_estate_df['Price'], 
         bins = 8,
         color = '#108A99')
plt.title("Distribution of Real Estate Prices", fontsize = 16, weight = 'bold')
plt.ylabel("Number of Properties")
plt.xlabel("Price in (000 $)")
sns.despine()
plt.show()
```
