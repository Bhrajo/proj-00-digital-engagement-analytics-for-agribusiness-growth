'startswith and endswith'

```python
# Checking if a string data truly starts with a word or a letter
print(s1.startswith('Cost'))
                            # case sensitive 
print(s1.startswith('cost')) 
                             # Any string character that starts can be checked
print(s.startswith('Pri'))
                             # It could be a morphem or even a letter
print(s.startswith('P'))
```

```python
# Checking if a string data truly ends with a word or a letter
print(s1.endswith('unit'))   # returns False if the string data does not end 
                             # with the specified word or letter
print(s1.endswith('cost'))
print(s1.endswith('nit'))
```

'strip' Method

```python
full_name = pd.Series(['Jeremiah Kekeli', 'Kofi Bentil', 
                       'Sandra Amenorvi', 'Godwin Nutor',
                       'Solomon Ametepe', 'Selorm Deyegbe'])
full_name
```

```python
#trying to take away some of the characters or words as specified per record
print(full_name[0].strip('J'))
print(full_name[1].strip('Bentil'))

# taking away words from left using the left-strip ('lstrip') method per record

print(full_name[2].lstrip)
print(full_name[3].rstrip('Godwin'))
print(full_name[4].rstrip('Ametepe'))
```

Using a String Accssor to strip words or characters in the whole Series

```python
print(full_name.str, '\n') 

print(full_name[0], '\n') # this prints the first indexed[0] record
                            # in the series

print(full_name.str[0]) # this strips away the first 
                            #character[0] in every record
```

```python
# Generating a new Series where the first words (first names) will be striped 
# throughout the whole Series leaving us a new Series (last_name)
last_name_split = full_name.str.split() # splits the string accessor into groups of lists
print(last_name_split, '\n')

last_name = full_name.str.split().str[1]
print(f"last_name: \n {last_name} \n")

first_name = full_name.str.split().str[0]
print(f"first_name: \n {first_name}")
```
