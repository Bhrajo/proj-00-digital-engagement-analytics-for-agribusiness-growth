#### **Saving files in NumPy** <br> np.save

```python
# Parsing a csv file with NumPy generate from text method (np.genfromtxt)
lending_co = np.genfromtxt('Lending-Company-Saving.csv', 
                           delimiter = ',', 
                           dtype = str)
print(lending_co)
```

```python
# Saving the generated text(the csv file that was parsed) with NumPy (np.save) as a ".npy" file type
np.save("Lending-Company-Saving", lending_co)
```

```python
# Loading the ".npy" file into a variable (lending_data_save) using Numpy (np.load)
lending_data_save = np.load("Lending-Company-Saving.npy")

print(lending_data_save)
```

```python
# Comparing the genrated ".csv" file to the loaded ".npy" file to see to if the contents are the same 
np.array_equal(lending_data_save, lending_co)
```

#### **Saving files in NumPy** <br> np.savez <br> The '.npz' can save multiple arrays of '.npy' files. It's like a suit/an archive/a collection of '.npy' files

```python
# Parsing a csv file with NumPy generate from text (np.genfromtxt) method 
lending_co = np.genfromtxt('Lending-Company-Saving.csv', 
                           delimiter = ',', 
                           dtype = str)

lending_data_save = np.load("Lending-Company-Saving.npy")
```

```python
# Saving the two '.npy' files (lending_co, and lending_data_save) in a .'npz. collection
np.savez("Lending-Company-Saving", lending_co, lending_data_save)
```

```python
lending_data_savez = np.load("Lending-Company-Saving.npz")

print(lending_data_savez['arr_1'])
```

```python
# Creating variable (company and data_save ) to store each file 
np.savez("Lending-Company-Saving", company =lending_co, data_save = lending_data_save)
```

```python
lending_data_savez = np.load("Lending-Company-Saving.npz") 
lending_data_savez.files
```

```python
print(lending_data_savez['data_save'])
```

Substituting missing values 

```python
lending_CDNN = np.genfromtxt(dataFile_NAN, delimiter = ';')
print(lending_CDNN,'\n')

## creating a mean value to substitute missing values using np.nanmean
sub_mean = np.nanmean(lending_CDNN, axis = 0).round(2)

## Creating a place holder for all missing values "temp_fill"
temp_fill  = np.nanmax(lending_CDNN).round(2) + 1 

## checking the mean of [0] with the sub_mean
print(f"sub_mean value: {sub_mean[0]}")
print(f"temp_fill value: {temp_fill}")
```

```python
lending_CDNN = np.genfromtxt(dataFile_NAN, delimiter = ';', filling_values = temp_fill)

print(lending_CDNN, '\n')
## The mean of [0] after applyting the temp_fill 
print()

## The impact of filling with temp_fill on the dataset 
print(f"The mean of [0] before the fill is {sub_mean[0]}\nThe mean after the fill is {np.mean(lending_CDNN[:,0]).round(2)}")
print(f"The difference is {abs(sub_mean[0] - np.mean(lending_CDNN[:,0]).round(2))}\nThe difference is too significant to overlook")
```

```python
## replaicing the temp_fill values with the sub_mean in the dataset
lending_CDNN[:,0] = np.where(lending_CDNN[:,0] == temp_fill,
                             sub_mean[0],
                             lending_CDNN[:,0])

print(np.mean(lending_CDNN[:,0]).round(2))
```

```python
## Using a for-loop to distribute the same result in the rest of the dataset 
for i in range(lending_CDNN.shape[1]):
    lending_CDNN[:,i] = np.where(lending_CDNN[:,i] == temp_fill, 
                                 sub_mean[i],
                                 lending_CDNN[:,i])                   
```
