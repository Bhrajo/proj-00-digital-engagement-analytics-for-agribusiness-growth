### **NumPy For Machine Learning**

```python
import numpy as np
np.version.full_version
```

```python
a = np.array([0,1,2,3,4,5])

a
```

```python
# A one dimensional (.ndim) array of 6 elements (.shape)
print(a.ndim)
print(a.shape)
```

We can now transform this array in to a 2D matrix

```python
# Reshaping the array into a two dimensional array of 3 rows and 2 columns (3,2)
aa1 = a.reshape(3,2)
aa1
```

```python
print(aa1.ndim)
print(aa1.shape)
```

```python
# Reshaping the array into a two dimensional array of 2 rows and 3 columns (2,3)

aa2 = a.reshape(2,3)
aa2
```

```python
print(aa2.ndim)
print(aa2.shape)
```

Preserving the true copy of a array

```python
## Only array "aa1" was altered but the change affected all others 
## The original content of all arrays were overighted
aa1[1][0] = 77
print(aa1)
print(aa2)
print(a)
```

```python
## To maintain the True copy the data the alteration should be done on a Copy (.copy())
ac = a.reshape((3,2)).copy()
ac
```

```python
# Here, c and a are totally independent copies because it's the copy that was altered
ac[0][0] = -99
print(a, '\n')

print(ac)
```

Another big advantage of NumPy arrays is that the operations are propagated
to the individual elements.

```python
print(a, )
a * 2
```

```python
print(a**2)

print(a > 4)
```

#### **Indexing** 


```python
# Using arrays as indices
print(a, '\n')

# Accessing the 2nd, 3rd, and fourth elements of array 'a' with and array of indices 
a[np.array([2,3,4])]
```

```python
print(a > 4)

## Accessing all the values greater than 4 in array 'a'
a[a > 4]
```

```python
## This can be used to even alter different elements of the dataset all at once
print(aa1)

## Using this to trim all outliers 
aa1[aa1 > 4] = 4
aa1
```

```python
## Triming can also be achieved by using the '.clip()' function
print(a)

a.clip(0,4)
```

Handling Non-existing values

```python
c = np.array([1, 2, np.nan, 3, 4]) # let's pretend we read this from a txt file
print(c)

np.isnan(c)
```

```python
print(c[~np.isnan(c)])

c[np.isnan(c)] = np.mean(c[~np.isnan(c)])

c
```
